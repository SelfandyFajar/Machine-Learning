```python
#machine learning
print("Nama: Selfandy Fajar Kautsara Rahutama \nNPM: 41155050210032")

import keras
import numpy as np

#model
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[2])])

#model compile
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
#set data
print("Nama: Selfandy Fajar Kautsara Rahutama \nNPM: 41155050210032")
xs = np.array([[1, 1], [1, 0], [0, 1],[0, 0]],dtype=int)
ys = np.array([1, 0, 0, 0],dtype=int)
```

```python
#Menampilkan arsitektur awal
print("Nama: Selfandy Fajar Kautsara Rahutama \nNPM: 41155050210032")
model.summary()
```

```python
#cek bobot model awal
print("Nama: Selfandy Fajar Kautsara Rahutama \nNPM: 41155050210032")
weights = model.get_weights()
weights
```

```python
#training
print("Nama: Selfandy Fajar Kautsara Rahutama \nNPM: 41155050210032")
model.fit(xs, ys, epochs=1000)
```

```python
#test
print("Nama: Selfandy Fajar Kautsara Rahutama \nNPM: 41155050210032")
data = [1,1]
answer = model.predict(np.array([data]))
print(answer)
```

```python
#training
print("Nama: Selfandy Fajar Kautsara Rahutama \nNPM: 41155050210032")
model2.fit(xs, ys, epochs=1000)
```
