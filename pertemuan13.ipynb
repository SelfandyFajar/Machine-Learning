```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

!pip install SimpSOM
import simpsom as sps
```

```python
print('Nama: Selfandy Fajar Kautsara Rahutama \nNPM: 41155050210032')

url = 'https://raw.githubusercontent.com/kokocamp/vlog119/refs/heads/main/vlog119.csv'
vlog139 = pd.read_csv(url)

X = vlog139[['gpa','gmat','work_experience','admitted']]
y = vlog139['admitted']

scaler = StandardScaler()
data = scaler.fit_transform(pd.DataFrame(X))
labels = scaler.fit_transform(pd.DataFrame(y))

print(data)
```

```python
print('Nama:Selfandy Fajar Kautsara Rahutama \nNPM: 41155050210032')
net = sps.SOMNet(50, 50, data)
train_algo = 'batch'
epochs = 1000
start_learning_rate = 0.0
net.train(train_algo=str(train_algo), epochs=epochs, start_learning_rate=start_learning_rate)

net.nodes_graph()
net.diff_graph()
net.project(data, labels=labels)
net.cluster(data)
```
